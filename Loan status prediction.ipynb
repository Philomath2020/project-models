```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Data ni DataFramega yuklash
df = pd.read_csv('/content/train_u6lujuX_CVtuZ9i (1).csv')
df.head()
```

```python
# ustun va qatorlar soni
df.shape
```

```python
# ustunlar bo'yicha umumiy malumotlar dtype , not-null
df.info()
```

```python
# Ustunlardagi mavjud bo'lmagan qiymatlar soni
df.isnull().sum()
```

```python
# Qiymat mavjud bo'lmagan qatorlarni tashlab yuborish
df = df.dropna()
```

```python
# Loan_Status ustunini sonli qiymatga aylantirish
df.replace({'Y':1,'N':0}, inplace=True)
```

```python
# ustundagi qiymatlarni ko'rish
df['Dependents'].value_counts()
```

```python
# qiymatlarni sonli qiymatlarga o'zgartirish

df.replace(to_replace='3+', value=4,inplace=True)

df.replace({'Married':{'Yes':1,'No':0}, 'Gender':{'Male':1,'Female':0}, 'Self_Employed':{'Yes':1,'No':0},
           'Property_Area':{'Urban':0,'Semiurban':1,'Rural':2}, 'Education':{'Graduate':1,'Not Graduate':0}}, inplace=True)

df['Dependents']=pd.to_numeric(df['Dependents'])
```

```python
df.info()
```

```python
# Malumotlarni ajratish

X = df.drop(columns=['Loan_ID','Loan_Status'], axis=1)
Y = df['Loan_Status']
```

```python
# train test qismlarga ajratish
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=2)
```

Support Vector Machine model -> Training

```python
SVM_model = svm.SVC(kernel='linear')
SVM_model.fit(X_train, Y_train)
```

**Model Evolution **

```python
X_train_prediction = SVM_model.predict(X_train)
Train_data_accuray = accuracy_score(X_train_prediction, Y_train)
print("Accuracy data score fl", Train_data_accuray)
```

```python
X_test_prediction = SVM_model.predict(X_test)
Test_data_accuray = accuracy_score(X_test_prediction, Y_test)
print("Accuracy data score fl", Test_data_accuray)
```
